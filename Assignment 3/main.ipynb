```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_blobs
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering

# --- DBSCAN ---
def dbscan_cluster(X, eps=0.3, min_samples=5):
    """Apply DBSCAN clustering."""
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(X)
    return labels

def plot_clusters(X, labels, title):
    """Plot clustering results."""
    plt.figure(figsize=(6,4))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# --- K-Means ---
def kmeans_cluster(X, n_clusters):
    """Apply K-Means clustering."""
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(X)
    return labels

# --- Hierarchical Clustering ---
def hierarchical_cluster(X, n_clusters):
    """Apply Agglomerative Hierarchical clustering."""
    hierarchical = AgglomerativeClustering(n_clusters=n_clusters)
    labels = hierarchical.fit_predict(X)
    return labels

# --- Main Execution ---
if __name__ == "__main__":
    # Dataset 1: make_moons
    X_moons, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
    # Dataset 2: make_blobs with varying densities
    X_blobs, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)

    # --- Clustering and Visualization for make_moons ---
    print("make_moons dataset:")
    labels = dbscan_cluster(X_moons, eps=0.3, min_samples=5)
    plot_clusters(X_moons, labels, title='DBSCAN on make_moons')

    labels = kmeans_cluster(X_moons, n_clusters=2)
    plot_clusters(X_moons, labels, title='K-Means on make_moons')

    labels = hierarchical_cluster(X_moons, n_clusters=2)
    plot_clusters(X_moons, labels, title='Hierarchical on make_moons')

    # --- Clustering and Visualization for make_blobs ---
    print("make_blobs dataset:")
    labels = dbscan_cluster(X_blobs, eps=0.6, min_samples=5)
    plot_clusters(X_blobs, labels, title='DBSCAN on make_blobs')

    labels = kmeans_cluster(X_blobs, n_clusters=3)
    plot_clusters(X_blobs, labels, title='K-Means on make_blobs')

    labels = hierarchical_cluster(X_blobs, n_clusters=3)
    plot_clusters(X_blobs, labels, title='Hierarchical on make_blobs')
```
